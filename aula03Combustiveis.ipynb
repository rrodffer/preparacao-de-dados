```python
# Dados foram obtidos em https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2016 (... até 2023)
# Nomes dos arquivos foram padronizados em  https://www.facom.ufu.br/~albertini/dados/pmqc/pmqc_AAAA_MM.csv

# De 2016/01 a 2018/12 usa formato https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2016/pmqc_AAAA_MM.[json,csv]
# https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2016/pmqc_2016_01.csv
# De 2019/01 a 2022/06 usa
# https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2019/2019-01-pmqc.csv
#https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2022/2022-06-pmqc.csv
# De 2022/07 a 2022/12 usa
#https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2022/pmqc_2022_07.csv
# De 2023/01 a 2023/11 usa
#https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2023/pmqc-01.csv
#https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/pmqc/2023/pmqc-11.csv

import pandas as pd

pmqc = pd.DataFrame()

for ano in range(2022, 2023): # existem dados de 01/2016 a 11/2023
  for mes in range(1, 13):

    if mes < 10:
      mes = '0'+str(mes)

    url = "https://www.facom.ufu.br/~albertini/dados/pmqc/pmqc_"+str(ano)+"_"+str(mes)+".csv"
    if ano == 2023 and mes == 22: break  # ainda não existe esse arquivo

    pmqc = pd.concat( [pmqc, pd.read_csv(url, sep = ';')] )
    print(pmqc.shape)
```

```python
combustiveis_distintos = pmqc['GrupoProduto'].drop_duplicates()

print("Combustíveis distintos: \n", combustiveis_distintos)
```

```python
status_conformidade = pmqc['Conforme'].drop_duplicates()

print("Status: \n", status_conformidade)
```

```python
import re

def normalizar_string(s):
    # Remove caracteres especiais usando expressão regular
    s_sem_especiais = re.sub(r'[^a-zA-Z0-9]', '', s)
    # Transforma em maiúsculas
    return s_sem_especiais.upper()

# Normalizar a coluna 'Nome'
pmqc['Município_Normalizado'] = pmqc['Município'].apply(normalizar_string)

print(pmqc['Município_Normalizado'])
```

```python
nome_municipio_normalizado = 'UBERLANDIA'
status_conforme = 'Não'

municipio_filtrado = pmqc.loc[(pmqc['Município_Normalizado'] == nome_municipio_normalizado) & (pmqc['Conforme'] == status_conforme)]


print(municipio_filtrado)
```

```python
combustivel_carros = pmqc.query('GrupoProduto == "Gasolina"')

medicoes_por_linha = combustivel_carros.pivot(index='IdNumeric', columns='Categoria', values='Aspecto')

medicoes_por_linha.head()


```

```python
import pandas as pd

# Exemplo de DataFrame
dados = {'Nome': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob'],
         'Categoria': ['A', 'B', 'A', 'B', 'A'],
         'Valor': [10, 20, 30, 40, 50]}

df = pd.DataFrame(dados)

# Aplicar a função pivot
df_pivot = df.pivot(index='Nome', columns='Ensaio', values='Valor')

# Exibir o DataFrame original
print("DataFrame Original:")
print(df)

# Exibir o DataFrame pivotado
print("\nDataFrame Pivotado:")
print(df_pivot)
```
